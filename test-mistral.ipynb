```python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

device = "cuda" # the device to load the model onto
```

```python
# model_name = '/kaggle/input/mistral/pytorch/7b-instruct-v0.1-hf/1'
model_name = 'mistralai/Mistral-7B-Instruct-v0.1'
model = AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.bfloat16,
        device_map="auto",
        trust_remote_code=True,
    )
```

```python
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
messages = [
    {"role": "user", "content": "Generate a question and answer pair about the sun. Format your response as just the question and the following answer. Do not denote the question or the answer with any prefix. Do not include Q:, A:, Question:, Answer:, or anything like that before the question or the answer."},
]
for i in range(20):
    encodeds = tokenizer.apply_chat_template(messages, return_tensors="pt")
    model_inputs = encodeds
    generated_ids = model.generate(model_inputs, max_new_tokens=1000, do_sample=True)
    decoded = tokenizer.batch_decode(generated_ids)
    print(' ')
    print(decoded[0])
```

```python

# Specify the file path
file_path = "output.txt"

# Open the file in write mode ('w')
with open(file_path, 'w') as file:
    # Write the string to the file
    file.write(decoded[0])

# print(f'String has been saved to {file_path}')
```
